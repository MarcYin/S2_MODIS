```python
vals = np.array([i.split('\t') for i in a.split('\n')]).astype(float)
```

```python
dat = [[np.arange(i[0]-2, i[0]+2, 1), np.arange(i[1]-3, i[1]+3, 1) ] for i in vals]
```

```python
fnames = []
for i in dat:
    lats, lons = i
    for lat0 in lats:
        for lon0 in lons:
            if lat0>=0:
                lat_name = 'N%02d'%int(lat0)
            elif lat0<0:
                lat_name = 'S%02d'%int(abs(lat0))
            else:
                'Wrong lat given, and float is expected!'
            if lon0>=0:
                lon_name = 'E%03d'%int(lon0)
            elif lon0<0:
                lon_name = 'W%03d'%int(abs(lon0))
            fnames.append( lat_name+lon_name+ '.hgt.zip')
```
