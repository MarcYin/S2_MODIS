```python
import numpy as np
%matplotlib inline
import pylab as plt
from scipy import signal
a = []
for i in range(16):
    data = np.zeros((50,50))
    data[:] = (i+1)*3
    a.append(data)    
v = []
for i in range(0,16,4):
    v1 = np.hstack((a[i],a[i+1], a[i+2], a[i+3]))
    v.append(v1) 
for i,j in enumerate(v):
    if i==0:
        a = j
    else:
        a = np.vstack((a,j))    
plt.imshow(a)
```

```python
m = np.arange(1,17)
m = m.reshape(4,4)
plt.imshow(m, interpolation='Nearest')
```

```python
inds = np.where(a)
indx = inds[0]+20
indy = inds[1]+10
for i,j in enumerate(indx):
    if j >=200:
        indx[i] = j-200
for i,j in enumerate(indy):
    if j >=200:
        indy[i] = j-200     
```

```python
ms = a[indx, indy].reshape(200,200)
```

```python
plt.imshow(ms)
```

```python
mas = np.zeros((50,50))
mask = (a==48)
w = 1./(1.*mask.sum())
mas[:] = w
conved = signal.fftconvolve(ms, mas, mode='valid')
conve = signal.fftconvolve(a, mas, mode='valid')
```

```python
minm0 = (abs(conve-48)).min()
np.where(abs(conve-48)==minm0)
```

```python
plt.imshow(abs(conved-48))
plt.colorbar()
```

```python
dif = abs(conved-48)
minm = np.min(dif)
inds = np.where(minm==dif)
```

```python
import numpy as np
%matplotlib inline
import pylab as plt
from scipy import signal

a = np.random.random((200,200))
```

```python
plt.imshow(a)
```

```python
m = (a[50:100, 50:100]).mean()
m
```

```python
mas = np.ones((50,50))
w = 1./(1.*mas.sum())
mas[:] = w
conved = signal.fftconvolve(a, mas, mode='valid')
```

```python
plt.imshow(conved)
plt.colorbar()
```

```python
dif = abs(conved - m)
minm = np.where(dif==dif.min())
```

```python
plt.imshow(abs(conved - m))
plt.plot(minm[1], minm[0], 'r+')
plt.xlim(0,150)
plt.ylim(150,0)
plt.colorbar()
```

```python
indx, indy = np.where(a)
```

```python
b = np.zeros_like(a)
indbx = indx - 10
indby = indy - 50
b[indx, indy] = a[indbx, indby]
plt.imshow(b)
```

```python
s_conved = signal.fftconvolve(b, mas, mode='valid')
```

```python
dif1 = abs(s_conved - m)
minm0 = np.where(dif1==dif1.min())

plt.imshow(dif1)
plt.plot(minm0[1], minm0[0], 'r+')
plt.xlim(0,150)
plt.ylim(150,0)
plt.colorbar()
```

```python
plt.imshow(abs(conved - m))
plt.plot(minm[1], minm[0], 'r+')
plt.xlim(0,150)
plt.ylim(150,0)
plt.colorbar()
```
