```python
data = [i[1] for i in vals]
```

```python
alpsf = []
for i in data:
    for ii in range(7):
        alpsf.append(i[ii][1][0])
```

```python
alpsf = np.array(alpsf).reshape(42,7,5)
```

```python
alpsf
```

```python
mask = [(alpsf[:,:,0]>23)&(alpsf[:,:,0]<25)&(alpsf[:,:,1]>300)&(alpsf[:,:,1]<330)]
```

```python
alpsf[:,:,0][mask] = alpsf[:,:,0][mask]*(2**0.5)
```

```python
alpsf[:,:,1][mask] = alpsf[:,:,1][mask]*(2**0.5)
```

```python
m = (alpsf[:,:,-1]<0)|(alpsf[:,:,-1]>100)
```

```python
m.shape
```

```python
selpsf = alpsf[~m]
```

```python
selpsf.shape
```

```python
selpsf = selpsf.reshape(38,7,5)
```

```python
names = ['B02', 'B03', 'B04', 'B08', 'B8A', 'B11', 'B12']
ylabels = ['Xstd', 'Ystd', 'Angle', 'Xshift', 'Yshift']
plot_config()
fig = plt.figure(figsize=(22,10))
mean = np.mean(selpsf, axis=0)
std = np.std(selpsf, axis=0)
for i in range(5):
    ax = fig.add_subplot(2,3,i+1)
    ax.plot(mean[:,i])
    ax.fill_between(range(7), mean[:,i]-3*std[:,i],mean[:,i]+3*std[:,i], color='0.8')
    ax.fill_between(range(7), mean[:,i]-std[:,i],mean[:,i]+std[:,i], color='0.5') 
    pretty_axes(ax)
    ax.set_yticks(np.linspace(int((mean[:,i]-3*std[:,i]).min()), int(1.2*((mean[:,i]+3*std[:,i]).max())), 5))
    ax.set_xticklabels(names)
    ax.set_ylabel(ylabels[i])
plt.tight_layout()
```
